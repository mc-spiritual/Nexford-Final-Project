```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('user_data.csv')
```

```python
print(df)
```

```python
#Convert all the float data type to integer for easy computation
df[df.select_dtypes(include='float').columns] = df.select_dtypes(include='float').astype(int)
```

```python
df.info()
```

```python
# Ages with Highest Income
top_ages = df.groupby('Age')['Income'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=top_ages.index, y=top_ages.values)
plt.title('Ages with Highest Average Income')
plt.ylabel('Average Income')
plt.savefig('top_ages_income.png')
plt.show()
plt.savefig('top_ages_income.png')
```

### Interpretation of the "Ages with Highest Average Income" Chart

The bar chart titled **"Ages with Highest Average Income"** visualizes the average income for each age group present in the dataset. From the chart, we observe the following:

- **Age 38** has the highest average income, followed closely by **age 43** and **age 42**. These age groups represent individuals in their early to mid-40s, suggesting that people in this age range tend to earn more on average.

Overall, the chart suggests a trend where income generally increases with age, peaking in the late 30s to early 40s, before stabilizing or declining slightly in later years. This pattern is consistent with typical career progression, where earnings grow with experience and seniority.

```python
# Group by gender and sum the expenses
gender_spend = df.groupby('Gender')[['Utilities', 'Entertainment', 'School Fees', 'Shopping', 'Healthcare']].sum()

# Plot
gender_spend.T.plot(kind='bar', figsize=(10,6))

plt.title('Gender Distribution Across Spending Categories', fontsize=16)
plt.ylabel('Total Spending (USD)', fontsize=14)
plt.xlabel('Spending Categories', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig("gender_spending_distribution_across_spending_categories.png")
plt.show()
```

### Interpretation of the "Gender Distribution Across Spending Categories" Chart

# The bar chart titled **"Gender Distribution Across Spending Categories"** compares total spending by gender across various categories: Utilities, Entertainment, School Fees, Shopping, and Healthcare.

**Key Observations:**
- **Females** spend significantly more on **Shopping** and **Entertainment** compared to males.
- **Males** have higher total spending on **School Fees** and **Utilities**.
- **Healthcare** spending is relatively similar between genders, with a slight edge for females.

**Conclusion:**  
Spending patterns differ by gender, with females allocating more to discretionary categories like Shopping and Entertainment, while males spend more on essential categories such as School Fees and Utilities. This insight can help tailor financial products or marketing strategies to different gender-based spending behaviors.
